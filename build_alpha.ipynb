Alphas


```python
def build_return(a):
    f_return = np.zeros(len(a))
    for i in range(1,len(a)):
        f_return[i] = (a[i] - a[i-1])/a[i-1]
    return (f_return)
```

```python
f_return=build_return(price_btc['pOpen'])
```

```python
alpha=np.zeros(len(price_btc)+1)
for d in range(5,len(price_btc)+1):
    #print((price_btc['pOpen'][d-5] - price_btc['pOpen'][d-1]) / price_btc['pOpen'][d-1])
    alpha[d]=(f_return[d-5] - f_return[d-1]) / f_return[d-1]
```

```python
def build_turnover(alpha_d):
    alpha_d=alpha_d.transpose()
    turnover=np.zeros(len(alpha_d))
    for i in range(1,len(alpha_d)):
    #print(alpha[i])
        turnover[i]=np.linalg.norm(alpha_d[i]-alpha_d[i-1], ord=1)*100
    plt.title('Turnover')    
    plt.plot(turnover)
    plt.show()
```

```python
build_turnover(alpha)
```

```python
build_alpha(alpha)
```

```python
def build_pnl_by_alpha_new(alpha,filt):
    pnl=np.zeros((len(alpha), 1))
    for i in range (5,len(filt)-1):
        #print((filt.transpose())[i+1],alpha[i])
        pnl[i]=alpha[i]*(filt.transpose())[i-1]
    #build_graph(pnl)
    return pnl
```

```python
pnl=build_pnl_by_alpha_new(alpha,f_return)
```

```python
plt.plot(pnl)
plt.show()
```
